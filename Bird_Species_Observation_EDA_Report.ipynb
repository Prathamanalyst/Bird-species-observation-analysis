# Bird Species Observation Analysis

## **Introduction**
Birds are vital indicators of ecosystem health, providing insights into environmental changes, biodiversity status, and habitat quality.  
This project focuses on analyzing **bird observation data** collected from **forest** and **grassland** habitats. The dataset includes detailed information on **species sightings, environmental conditions, observation times, and locations**, making it suitable for both ecological research and conservation planning.

### **Project Objectives**
- To identify **temporal patterns** in bird sightings (seasons, months, time of day).
- To explore **spatial biodiversity** patterns across habitats and plots.
- To study the relationship between **environmental factors** (temperature, humidity, wind, sky) and bird activity.
- To highlight **species of conservation concern** using watchlist and stewardship status.
- To provide **actionable insights** for wildlife conservation and habitat management.

### **Source of Data**
The dataset was provided by the company as part of a bird monitoring project.  
It consolidates multiple observation sheets from forest and grassland regions into a single cleaned dataset for analysis.

# Dataset Overview

The Bird Species Observation dataset contains detailed records of bird sightings collected from **forest** and **grassland** habitats.  
The aim is to analyze biodiversity patterns, environmental influences, and temporal trends to assist in ecological research and conservation efforts.

## **Column Descriptions**

| Column Name | Data Type | Description |
|-------------|-----------|-------------|
| **Admin_Unit_Code** | Categorical | Code of the administrative unit where the observation was made (e.g., ANTI, HAWK). |
| **Plot_Name** | String | Name of the specific observation plot within the habitat. |
| **Location_Type** | Categorical | Habitat type — mainly **Forest** or **Grassland**. |
| **Year** | Datetime | Year in which the observation was recorded. |
| **Date** | Date | Exact date of the observation. |
| **Start_Time** | Time | Time when the observation started. |
| **End_Time** | Time | Time when the observation ended. |
| **Observer** | Categorical | Name or code of the person who recorded the observation. |
| **Visit** | Integer | Visit number for the same plot within a monitoring cycle. |
| **Interval_Length** | Categorical | Length of the observation interval (e.g., 3 minutes, 5 minutes). |
| **ID_Method** | Categorical | Method used to identify the species (e.g., visual, auditory). |
| **Distance** | Categorical | Estimated distance from the observer to the bird (e.g., 0-25m, 25-50m). |
| **Flyover_Observed** | Boolean | Indicates if the bird was flying over the area during observation. |
| **Sex** | Categorical | Sex of the bird (Male, Female, Unknown). |
| **Common_Name** | String | Common name of the bird species (e.g., Northern Cardinal). |
| **Scientific_Name** | String | Scientific (Latin) name of the bird species. |
| **AcceptedTSN** | Integer | Taxonomic Serial Number for the species — a unique ID in taxonomic databases. |
| **AOU_Code** | String | American Ornithological Union code for the species. |
| **PIF_Watchlist_Status** | Boolean | Indicates if the species is listed on the Partners in Flight Watchlist (True/False). |
| **Regional_Stewardship_Status** | Boolean | Indicates if the species is a regional stewardship concern (True/False). |
| **Temperature** | Float | Temperature during the observation, in degrees Celsius. |
| **Humidity** | Float | Humidity during the observation, in percentage. |
| **Sky** | Categorical | Sky conditions (e.g., Clear, Partly Cloudy, Overcast). |
| **Wind** | Categorical | Wind conditions (e.g., Calm, Light breeze). |
| **Disturbance** | Categorical | Disturbance factors present (e.g., No effect, Slight effect). |
| **Initial_Three_Min_Cnt** | Boolean | Indicates if the bird was detected within the first three minutes of observation. |
| **Month** | Integer | Month number extracted from the observation date (1–12). |
| **Season** | Categorical | Season derived from the observation date (Spring, Summer, Fall, Winter). |
| **Weekday** | Categorical | Day of the week (Monday–Sunday). |

## 1. Temporal Analysis

This step focuses on understanding **when** bird sightings occur by analyzing:

- **Seasonal Trends** → Which seasons have the most sightings.
- **Monthly Patterns** → Fluctuations in bird activity across months.
- **Yearly Trends** → How sightings change over years.
- **Time of Day Patterns** → Hours with peak bird activity.

We will use **bar charts, line charts, and histograms** to visualize these patterns.


## 1.1 Seasonal Sightings

**Purpose:**  
To analyze bird observation counts across different seasons (e.g., Spring, Summer, etc.) to identify patterns in bird activity.

**Approach:**  
- Group dataset by the `Season` column
- Count total observations per season
- Plot a bar chart to compare seasons

**Chart Type:** Bar Chart


```python
# Seasonal observation counts
season_counts = df['Season'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=season_counts.index, y=season_counts.values, color="skyblue")
plt.title("Bird Observations by Season")
plt.xlabel("Season")
plt.ylabel("Number of Observations")

# Add value labels on top of each bar
for i, val in enumerate(season_counts.values):
    plt.text(i, val + 50, str(val), ha='center', fontsize=10)

plt.show()
```

**Insights:**  
- Highest observations were recorded in Summer (10,503 sightings).

- Lowest observations occurred in Spring (4,863 sightings).

**This suggests that bird activity is significantly higher in warmer months.**


## 1.2 Monthly Sightings

**Purpose:**  
To see how bird sightings vary month-to-month, helping identify peak observation periods.

**Approach:**  
- Group dataset by `Month`
- Count total observations per month
- Use a line chart to show trends across months

**Chart Type:** Line Chart


```python
month_counts = df['Month'].value_counts().sort_index()
plt.figure(figsize=(8,4))
sns.barplot(x=month_counts.index, y=month_counts.values, color="lightgreen")
plt.title("Bird Observations by Month")
plt.xlabel("Month")
plt.ylabel("Number of Observations")
for i, val in enumerate(month_counts.values):
    plt.text(i, val + 30, str(val), ha='center', fontsize=9)
plt.show()
```

**Insights:**  
- June had the highest number of sightings (6,207), followed by May.

- July had the lowest number of sightings (4,296).

**This could indicate peak migration or breeding during early summer.**


## 1.3 Yearly Sightings

**Purpose:**  
To examine observation counts across different years to detect any trends or anomalies over time.

**Approach:**  
- Group dataset by `Year`
- Count observations for each year
- Use a bar chart for year-to-year comparison

**Chart Type:** Bar Chart


```python
year_counts = df['Year'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=year_counts.index, y=year_counts.values, color="orange")
plt.title("Bird Observations by Year")
plt.xlabel("Year")
plt.ylabel("Number of Observations")
for i, val in enumerate(year_counts.values):
    plt.text(i, val + 20, str(val), ha='center', fontsize=10)
plt.show()
```

**Insights:**  
- **Data is from a single year (2018) with 15,366 total records.**


## 1.4 Observation Time Patterns

**Purpose:**  
To find out which times of day have more bird activity, possibly indicating feeding or migration patterns.

**Approach:**  
- Convert `Start_Time` to hours (integer)
- Count observations by hour
- Plot a histogram to visualize activity peaks

**Chart Type:** Histogram


```python
# Time of Day Patterns
df['Hour'] = df['Start_Time'].str.split(':').str[0]   # Extract hour part as string
df['Hour'] = pd.to_numeric(df['Hour'], errors='coerce')  # Convert to number for plotting

hour_counts = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10,4))
sns.lineplot(x=hour_counts.index, y=hour_counts.values, marker='o', color="purple")
plt.title("Bird Observations by Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Observations")
for x, y in zip(hour_counts.index, hour_counts.values):
    plt.text(x, y + 5, str(y), ha='center', fontsize=10)
plt.show()
```

**Insights:**  
- Bird activity peaked at 7 AM (4,014 sightings).

- The lowest recorded activity was at 10 AM (1,021 sightings).

**Early morning appears to be the most active time for bird sightings.**

**Insights:**  


- The top three plots (**ANTI-0105**, **MANA-0047**, **MONO-0057**) each recorded **27 unique species**.

## 3.3 Activity Patterns by Interval Length

## 3.4 Sex Ratio per Species (Top 5 Species)

**Purpose:**  
To understand the male-to-female observation ratio for the most common species.

**Approach:**  
- Take the top 5 most common species
- Group by `Common_Name` and `Sex`
- Plot a stacked bar chart with data labels

**Chart Type:** Stacked Bar Chart


```python
# Top 5 species for sex ratio
top5_species = df['Common_Name'].value_counts().head(5).index
sex_ratio_df = (df[df['Common_Name'].isin(top5_species)].groupby(['Common_Name', 'Sex'], observed=False).size().reset_index(name='Count'))

plt.figure(figsize=(10,5))
ax = sns.barplot(data=sex_ratio_df, x='Common_Name', y='Count', hue='Sex', palette="Set3")

plt.title("Sex Ratio for Top 5 Bird Species")
plt.xlabel("Species")
plt.ylabel("Number of Observations")

# Add data labels above bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

**Insights:**
- Most birds have undetermined sex: Carolina Wren (993), Northern Cardinal (1036).

- Indigo Bunting has many more males (450) than females (6).

- Eastern Tufted Titmouse (713) and Red-eyed Vireo (692) are mostly undetermined, with few males (7 and 44) and almost no females.

## 4: Environmental Conditions

## 4.1 Temperature Distribution

**Purpose:**  
To understand the range and frequency of temperatures during bird observations. This helps determine if certain temperatures correspond with higher observation counts.

**Approach:**  
- Use the `Temperature` column.  
- Plot a histogram with bins to visualize distribution.  

**Chart Type:** Histogram


```python
plt.figure(figsize=(12,5))
counts, bins, patches = plt.hist(df['Temperature'], bins=20, color='skyblue', edgecolor='black')
plt.title('Temperature Distribution During Observations')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')

# Add data labels
for count, bin_edge in zip(counts, bins):
    if count > 0:
        plt.text(bin_edge + (bins[1]-bins[0])/2, count, str(int(count)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

**Insights:**  
- Observation temperatures ranged from 11°C to 37.3°C, with most between 19.7°C and 25°C.

- Median temperature was 22.2°C, showing surveys favored mild weather.

- Very few observations occurred above 30°C, suggesting lower activity or fewer surveys in hotter conditions.
