```python
# Read and combine all sheets from Forest Excel file

import pandas as pd

# Define the file name
forest_file = "Bird_Monitoring_Data_FOREST.XLSX"

# Function to read all sheets from the Forest Excel file
def read_forest_sheets(file_path):
    # Load the Excel workbook
    xl = pd.ExcelFile(file_path)

    # List to collect DataFrames from each sheet
    forest_sheets = []

    # Loop through each sheet name in the file
    for sheet in xl.sheet_names:
        # Read the sheet into a DataFrame
        df = xl.parse(sheet)

        # Add identifying columns
        df['Admin_Unit_Code'] = sheet           # From sheet name (e.g., ANTI, CATO)
        df['Location_Type'] = 'Forest'          # Tag this as forest data

        # Add this DataFrame to the list
        forest_sheets.append(df)

    # Combine all the DataFrames into one
    forest_data = pd.concat(forest_sheets, ignore_index=True)

    return forest_data

# Run the function and store the result in a variable
forest_df = read_forest_sheets(forest_file)

# Show the shape and preview first few rows
print("Forest data combined successfully!")
print("Shape of forest_df:", forest_df.shape)
forest_df.head()
```

```python
# Read and combine all sheets from Grassland Excel file

# Define the file name
grassland_file = "Bird_Monitoring_Data_GRASSLAND.XLSX"

# Function to read all sheets from the Grassland Excel file
def read_grassland_sheets(file_path):
    # Load the Excel workbook
    xl = pd.ExcelFile(file_path)

    # List to collect DataFrames from each sheet
    grassland_sheets = []

    # Loop through each sheet name in the file
    for sheet in xl.sheet_names:
        # Read the sheet into a DataFrame
        df = xl.parse(sheet)

        # Add identifying columns
        df['Admin_Unit_Code'] = sheet            # From sheet name
        df['Location_Type'] = 'Grassland'        # Tag this as grassland data

        # Add this DataFrame to the list
        grassland_sheets.append(df)

    # Combine all the DataFrames into one
    grassland_data = pd.concat(grassland_sheets, ignore_index=True)

    return grassland_data

# Run the function and store the result
grassland_df = read_grassland_sheets(grassland_file)

# Show the shape and preview of the grassland data
print("Grassland data combined successfully!")
print("Shape of grassland_df:", grassland_df.shape)
grassland_df.head()
```

```python
# Combine forest and grassland data into one dataset

# Combine both DataFrames using concat
combined_df = pd.concat([forest_df, grassland_df], ignore_index=True)

# Show shape and preview
print("Combined DataFrame created successfully!")
print("Shape of combined_df:", combined_df.shape)

# View the first few rows
combined_df.head()
```

```python
# Save the combined data to a CSV file

# Save the combined DataFrame as a CSV file
combined_df.to_csv("Combined_Bird_Data.csv", index=False)

print("File saved successfully as 'Combined_Bird_Data.csv'")
```

```python
# Check basic structure and missing values

print("Dataset Info:")
combined_df.info()

print("Missing Values:")
missing = combined_df.isnull().sum()
print(missing[missing > 0])

print("Summary (Numeric):")
display(combined_df.describe())

print("Summary (Categorical):")
display(combined_df.describe(include='object'))
```

### Drop Columns with High Missing Values

We dropped the following columns due to high missing values (50%+):
- `Sub_Unit_Code`
- `Site_Name`
- `NPSTaxonCode`
- `TaxonCode`
- `Previously_Obs`

This helps clean the dataset without sacrificing useful data.

```python
columns_to_drop = ['Sub_Unit_Code', 'Site_Name', 'NPSTaxonCode', 'TaxonCode', 'Previously_Obs']

combined_df.drop(columns=columns_to_drop, axis=1, inplace=True)

print("Dropped columns with too much missing data.")
print("New shape:", combined_df.shape)
```

### Handle Remaining Missing Values

We cleaned the remaining missing values with the following strategies:

- Dropped 2 rows missing `ID_Method` (essential for analysis)
- Filled:
  - `Distance` → "Unknown"
  - `Sex` → "Undetermined"
  - `AcceptedTSN` → -1 (numeric placeholder)

This ensures a clean and consistent dataset for further exploration.


```python
# Drop rows with missing ID_Method (only 2 rows)
combined_df = combined_df.dropna(subset=['ID_Method'])

# Fill missing Distance with "Unknown"
combined_df['Distance'] = combined_df['Distance'].fillna('Unknown')

# Fill missing Sex with "Undetermined"
combined_df['Sex'] = combined_df['Sex'].fillna('Undetermined')

# Fill missing AcceptedTSN with -1 (or another placeholder if you want)
combined_df['AcceptedTSN'] = combined_df['AcceptedTSN'].fillna(-1)

# Check again for any remaining missing values
combined_df.isnull().sum()[combined_df.isnull().sum() > 0]
```

### Remove Duplicate Rows

We removed exact duplicate records to avoid bias and double-counting.

This ensures that each bird observation is unique and not overrepresented.


```python
# Check number of duplicate rows
duplicates = combined_df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
# Drop duplicates and keep the first occurrence
combined_df = combined_df.drop_duplicates()

# Confirm new shape
print("New shape after dropping duplicates:", combined_df.shape)
```

###  Add Date-Based Features

We extracted:
- `Month` from the `Date` column
- `Season` based on the month (Spring, Summer, Autumn, Winter)
- `Weekday` to explore trends across days of the week

These features will help in visualizing temporal patterns in bird observations.


```python
# Extract Month from Date
combined_df['Month'] = combined_df['Date'].dt.month

# Map Month to Season (Spring, Summer)
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

combined_df['Season'] = combined_df['Month'].apply(get_season)

# Extract Weekday (e.g., Monday, Tuesday)
combined_df['Weekday'] = combined_df['Date'].dt.day_name()

# Preview
combined_df[['Date', 'Month', 'Season', 'Weekday']].head()
```

### Standardize Categorical Columns

To ensure clean analysis, we standardized key categorical columns:

- `Sex` → Fixed case and combined values like "m", "male", "M" into "Male"
- `Distance` → Cleaned casing and filled missing as "Unknown"
- `Sky`, `Wind`, `Disturbance` → Converted to title case and stripped spaces

This prevents duplicated categories and improves clarity during visualization.


```python
# Standardize 'Sex' column
combined_df['Sex'] = combined_df['Sex'].str.strip().str.lower()
combined_df['Sex'] = combined_df['Sex'].replace({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female',
    'undetermined': 'Undetermined',
    'u': 'Undetermined',
    '': 'Undetermined'
})
combined_df['Sex'] = combined_df['Sex'].fillna('Undetermined')

# Standardize 'Distance'
combined_df['Distance'] = combined_df['Distance'].str.strip().str.title()
combined_df['Distance'] = combined_df['Distance'].replace({
    '': 'Unknown'
})
combined_df['Distance'] = combined_df['Distance'].fillna('Unknown')

# Standardize 'Sky', 'Wind', 'Disturbance' – clean casing/whitespace
for col in ['Sky', 'Wind', 'Disturbance']:
    combined_df[col] = combined_df[col].str.strip().str.title()

# Preview changes
combined_df[['Sex', 'Distance', 'Sky', 'Wind', 'Disturbance']].drop_duplicates().sort_values(by='Sex').head(10)
```

### Convert Categorical Columns to `category` Type

We converted selected columns to `category` type for performance and consistency. This includes:

- Observation context: `Sex`, `Distance`, `Interval_Length`, `Sky`, `Wind`, `Disturbance`
- Location/group IDs: `Admin_Unit_Code`, `Observer`, `Visit`
- Time-based: `Season`, `Month`, `Weekday`
- Binary-like fields: `Flyover_Observed`, `PIF_Watchlist_Status`, etc.

This ensures better performance during EDA and modeling.


```python
# Columns with few unique values — good for conversion
cat_cols = ['Location_Type', 'Flyover_Observed', 'Regional_Stewardship_Status',
    'Initial_Three_Min_Cnt', 'Season', 'PIF_Watchlist_Status',
    'Observer', 'ID_Method', 'Distance', 'Visit', 'Month', 'Sex',
    'Interval_Length', 'Disturbance', 'Wind', 'Sky', 'Weekday', 'Admin_Unit_Code']

# Convert them to category type
for col in cat_cols:
    if col in combined_df.columns:
        combined_df[col] = combined_df[col].astype('category')

# Check if it worked
combined_df.dtypes[combined_df.dtypes == 'category']
```
